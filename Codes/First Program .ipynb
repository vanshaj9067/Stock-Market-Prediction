```python
google_data.info()
```

```python
x = google_data[['Open','High','Low','Volume']]
y = google_data['Close']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
x_scale = pd.DataFrame(x_scaler,columns=['Open','High','Low','Volume'])
x_scale
```

```python
#linear regression 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
xtrain,xtest,ytrain,ytest = train_test_split(x_scale,y,test_size=0.2)
modellr = LinearRegression()
modellr.fit(xtrain,ytrain)
ypred = modellr.predict(xtest)
r2s = r2_score(ytest,ypred)
print(r2s)
```

```python
#decision tree regression 
from sklearn.tree import DecisionTreeRegressor
modeldtr = DecisionTreeRegressor(splitter="random",max_depth=9)
modeldtr.fit(xtrain,ytrain)
ypred2=modeldtr.predict(xtest)
r2s2 = r2_score(ytest,ypred2)
print(r2s2)
```

```python
#svr
from sklearn.svm import SVR
modelsvr = SVR(C=1.0, epsilon=0.2)
modelsvr.fit(xtrain,ytrain)
ypred3=modelsvr.predict(xtest)
r2s3=r2_score(ytest,ypred3)
print(r2s3)
```

```python
#voting ensemble 
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import VotingRegressor
estimators = [('lr',modellr),('dtr',modeldtr),('svr',modelsvr)]
vr = VotingRegressor(estimators=estimators)
xt = cross_val_score(vr,x_scale,y,cv=10,scoring='r2')
print(np.round(np.mean(xt),2))
```

we can see that our voting ensemble model have a negative r2 score which is worse so we go through with linear regression

```python
TCS_data.info()
```

```python
xtcs = TCS_data[['Open','High','Low','Volume']]
ytcs = TCS_data['Close']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xt_scaler = scaler.fit_transform(xtcs)
xt_scale = pd.DataFrame(x_scaler,columns=['Open','High','Low','Volume'])
xt_scale
```

```python
#linear regression 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
xtrainc,xtestc,ytrainc,ytestc = train_test_split(xt_scale,ytcs,test_size=0.2)
modellr = LinearRegression()
modellr.fit(xtrainc,ytrainc)
ypredc = modellr.predict(xtestc)
r2sc = r2_score(ytestc,ypredc)
print(r2sc)
```

```python
#decision tree regression 
from sklearn.tree import DecisionTreeRegressor
modeldtr = DecisionTreeRegressor(splitter="random",max_depth=9)
modeldtr.fit(xtrainc,ytrainc)
ypred2c=modeldtr.predict(xtestc)
r2s2c = r2_score(ytestc,ypred2c)
print(r2s2c)
```

```python
#svr
from sklearn.svm import SVR
modelsvr = SVR(C=1.0, epsilon=0.2)
modelsvr.fit(xtrainc,ytrainc)
ypred3c=modelsvr.predict(xtestc)
r2s3c=r2_score(ytestc,ypred3c)
print(r2s3c)
```

We see that Decison tree regressor works good on tcs data
