## Stock Market Overview

The stock market is a complex system where investors buy and sell shares of publicly traded companies. Understanding the fundamental concepts and terminologies is essential for navigating and analyzing market activities effectively.

### 1. **What is the Stock Market?**
- **Definition**: A platform where investors can buy and sell shares of publicly listed companies.
- **Components**:
  - **Stock Exchanges**: Physical or electronic venues where trading occurs (e.g., NYSE, NASDAQ, BSE).
  - **Listed Companies**: Businesses that have issued shares to the public through an initial public offering (IPO).
  - **Investors**: Individuals or institutions that purchase shares to gain ownership stakes and potential returns.

### 2. **Key Terminologies**

#### a. **Stock (Share)**
- **Definition**: A unit of ownership in a company.
- **Types**:
  - **Common Stock**: Provides voting rights and potential dividends.
  - **Preferred Stock**: Typically no voting rights but has a higher claim on assets and dividends.

#### b. **Ticker Symbol**
- **Definition**: A unique series of letters representing a publicly traded company's stock.
- **Example**: 
  - **GOOG** for Alphabet Inc. (Google)
  - **TCS** for Tata Consultancy Services
  - **BSESN** for BSE Sensex (index)

#### c. **Stock Exchange**
- **Definition**: An organized market where securities are bought and sold.
- **Examples**:
  - **New York Stock Exchange (NYSE)**
  - **NASDAQ**
  - **Bombay Stock Exchange (BSE)**

#### d. **Index**
- **Definition**: A statistical measure representing the performance of a group of stocks.
- **Examples**:
  - **BSE Sensex (BSESN)**: Tracks 30 major companies on the BSE.
  - **S&P 500**: Tracks 500 large-cap U.S. companies.
  - **NASDAQ Composite**: Includes over 3,000 stocks listed on NASDAQ.

#### e. **Bid and Ask Price**
- **Bid Price**: The highest price a buyer is willing to pay for a stock.
- **Ask Price**: The lowest price a seller is willing to accept for a stock.
- **Spread**: The difference between the bid and ask prices.

#### f. **Market Order vs. Limit Order**
- **Market Order**: An order to buy or sell immediately at the current market price.
- **Limit Order**: An order to buy or sell at a specific price or better.

#### g. **Volume**
- **Definition**: The number of shares traded during a specific period.
- **Significance**: Indicates the activity level and liquidity of a stock.

#### h. **Volatility**
- **Definition**: A measure of how much a stock's price fluctuates over a given period.
- **Types**:
  - **High Volatility**: Large price swings, indicating higher risk.
  - **Low Volatility**: Smaller price movements, indicating lower risk.

#### i. **Liquidity**
- **Definition**: The ease with which a stock can be bought or sold without affecting its price.
- **High Liquidity**: Stocks can be traded quickly with minimal price impact.
- **Low Liquidity**: Trades may significantly impact the stock price.

#### j. **Dividend**
- **Definition**: A portion of a company's earnings distributed to shareholders.
- **Types**:
  - **Cash Dividend**: Paid in cash.
  - **Stock Dividend**: Paid in additional shares.

### 3. **How the Stock Market Works**

- **Initial Public Offering (IPO)**:
  - **Definition**: The process by which a private company offers shares to the public for the first time.
  - **Purpose**: To raise capital for expansion, pay off debt, or provide liquidity to existing shareholders.

- **Secondary Market**:
  - **Definition**: The market where previously issued securities are traded among investors.
  - **Function**: Provides liquidity and enables price discovery.

- **Trading Mechanism**:
  - **Order Matching**: Buy and sell orders are matched through an exchange's order book.
  - **Execution**: Once matched, the trade is executed, and ownership of the shares is transferred.

### 4. **Investment Strategies**

#### a. **Fundamental Analysis**
- **Definition**: Evaluating a company's financial health, performance, and intrinsic value.
- **Components**:
  - **Financial Statements**: Income statement, balance sheet, cash flow statement.
  - **Ratios**: Price-to-earnings (P/E), debt-to-equity, return on equity (ROE).
  - **Qualitative Factors**: Management quality, competitive advantage, industry conditions.

#### b. **Technical Analysis**
- **Definition**: Analyzing statistical trends from trading activity, such as price and volume.
- **Tools**:
  - **Charts**: Line charts, bar charts, candlestick charts.
  - **Indicators**: Moving averages, Relative Strength Index (RSI), MACD.

#### c. **Quantitative Analysis**
- **Definition**: Using mathematical and statistical models to evaluate securities.
- **Approach**:
  - **Algorithmic Trading**: Automated trading based on predefined criteria.
  - **Risk Management Models**: Assessing and mitigating financial risks.

### 5. **Risk and Return**

- **Risk**: The possibility of losing some or all of the invested capital.
- **Return**: The gain or loss generated by an investment.
- **Risk-Return Tradeoff**: Higher potential returns typically come with higher risk.

### 6. **Regulatory Bodies**

- **Securities and Exchange Commission (SEC)**: U.S. regulatory body overseeing securities markets.
- **Securities and Exchange Board of India (SEBI)**: Regulatory authority for securities markets in India.
- **Role**:
  - **Protecting Investors**: Ensuring transparency and fairness in the markets.
  - **Regulating Exchanges**: Setting rules and guidelines for trading activities.
  - **Preventing Fraud**: Detecting and penalizing fraudulent activities.

### 7. **Market Participants**

- **Retail Investors**: Individual investors buying and selling securities for personal accounts.
- **Institutional Investors**: Entities like mutual funds, pension funds, and insurance companies managing large portfolios.

### 8. **Bull and Bear Markets**: 
   - A **bull market** refers to a phase where stock prices are rising or are expected to rise, signaling optimism and confidence among investors.
   - A **bear market** describes a situation where prices are falling or are expected to fall, reflecting pessimism and caution in the market. Investors typically take long positions during bull markets and short positions during bear markets, depending on their market outlook.

## Dataset Description

This dataset contains historical stock market data for three companies: **TCS**, **Google (GOOG)**, and the **BSE Sensex (BSESN)**. ## Dataset Fields

This section provides an in-depth explanation of each field in the stock market dataset. Understanding these fields is crucial for effectively analyzing and modeling stock price movements.

### 1. **Date**
- **Definition**: The specific calendar day when the trading activity occurred.
- **Details**:
  - **Format**: Typically in `YYYY-MM-DD` format (e.g., 2024-04-27).
  - **Importance**: 
    - Helps in time-series analysis.
    - Allows for the tracking of stock performance over time.
    - Essential for identifying trends, seasonal patterns, and cyclical behaviors in stock prices.
  - **Usage**:
    - Aligning data from different sources.
    - Performing time-based aggregations (e.g., daily, weekly, monthly analysis).

### 2. **Open**
- **Definition**: The price at which a stock begins trading when the market opens for the day.
- **Details**:
  - **Determination**: Based on the first trade executed during the trading session.
  - **Influencing Factors**:
    - Overnight news and events.
    - Market sentiment before the market opens.
    - Pre-market trading activities.
  - **Importance**:
    - Serves as a reference point for the day's trading activity.
    - Used in technical analysis to identify opening trends and patterns.

### 3. **High**
- **Definition**: The highest price at which the stock traded during the trading day.
- **Details**:
  - **Measurement**: Recorded as the peak price reached from the market open to close.
  - **Significance**:
    - Indicates the maximum buying interest during the day.
    - Helps assess the stock's volatility and price range.
  - **Usage**:
    - Identifying resistance levels in technical analysis.
    - Comparing with previous highs to gauge momentum.

### 4. **Low**
- **Definition**: The lowest price at which the stock traded during the trading day.
- **Details**:
  - **Measurement**: Recorded as the lowest price reached from the market open to close.
  - **Significance**:
    - Indicates the maximum selling pressure during the day.
    - Helps assess the stock's volatility and price range.
  - **Usage**:
    - Identifying support levels in technical analysis.
    - Comparing with previous lows to gauge bearish momentum.

### 5. **Close**
- **Definition**: The final price at which the stock trades when the market closes for the day.
- **Details**:
  - **Determination**: Based on the last executed trade before the market closes.
  - **Importance**:
    - Widely regarded as the most important price of the day.
    - Used as a benchmark for performance evaluation.
  - **Usage**:
    - Calculating daily returns.
    - Forming the basis for many technical indicators (e.g., moving averages).

### 6. **Adj Close (Adjusted Close)**
- **Definition**: The closing price adjusted for corporate actions such as dividends, stock splits, and rights offerings.
- **Details**:
  - **Adjustment Factors**:
    - **Dividends**: Cash payments made to shareholders.
    - **Stock Splits**: Increase in the number of shares while reducing the price proportionally.
    - **Rights Offerings**: Issuing additional shares to existing shareholders.
  - **Significance**:
    - Provides a more accurate reflection of the stock's value over time.
    - Essential for long-term performance analysis and comparison.
  - **Usage**:
    - Calculating historical returns without distortions from corporate actions.
    - Enhancing the reliability of technical and fundamental analysis.

### 7. **Volume**
- **Definition**: The total number of shares traded during the trading day.
- **Details**:
  - **Measurement**: Sum of all shares bought and sold throughout the day.
  - **Significance**:
    - Indicates the level of activity and liquidity in the stock.
    - High volume often accompanies significant price movements.
  - **Usage**:
    - Confirming trends (e.g., high volume on an uptrend suggests strength).
    - Identifying potential reversals or breakouts.

### Target

The goal is to predict the **Closing Price** based on the **Open**, **High**, and **Low** prices. Additionally, the company data will be used as an input feature to improve the learning capability of the model. The dataset combines stock data from **TCS**, **Google**, and **BSE Sensex** to allow the model to generalize better across different companies.

## Explanation of Each Field

1. **Open**: This is the price at which a stock starts trading when the market opens for that day. It is influenced by factors like overnight news, demand, and global events that occurred when the market was closed.

2. **High**: The highest price the stock reaches during a single trading day. This reflects the peak point of buying interest and market optimism for that day.

3. **Low**: The lowest price the stock reaches during the trading day, which signifies the lowest point of market sentiment and selling pressure.

4. **Close**: The last price at which the stock trades during the day. This is often considered the most important price, as it reflects the final market value for the day.

5. **Adj Close (Adjusted Close)**: This price is adjusted to account for corporate actions like dividends, stock splits, and new stock issuance. It gives a more accurate picture of the stock's true market value, especially for long-term analysis.

6. **Volume**: This represents the number of shares traded throughout the day. High trading volumes indicate strong interest in the stock, while low volumes can suggest weak demand.

### Importance of Open, High, Low, and Close Prices:

- **Close Price**: Generally considered the most important price because it reflects the final trading price of the day and is often used in technical analysis and financial reporting.
  
- **High and Low Prices**: Provide insights into the price volatility and the range of prices during the trading day. They can help identify support and resistance levels.
  
- **Open Price**: Indicates the starting point of trading for a given day and can provide context for the day's price movements.

```python
# Function to convert date column to dd-mm-yyyy format
def convert_date_format(df):
    # Convert the date format
    df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d').dt.strftime('%d-%m-%Y')
    df['Date'] = pd.to_datetime(df.Date, format='%d-%m-%Y')
    
    return df['Date']

# Print confirmation
print("Date column converted to dd-mm-yyyy format for BSESN:\n")

# Convert the date format for the BSESN DataFrame
preprocessed_df_bsesn['Date'] = convert_date_format(preprocessed_df_bsesn)

print("BSESN Dates:\n", preprocessed_df_bsesn['Date'])

# Print confirmation
print("Date column converted to dd-mm-yyyy format for Google:\n")

# Convert the date format for the Google DataFrame
preprocessed_df_goog['Date'] = convert_date_format(preprocessed_df_goog)

print("\nGoogle Dates:\n", preprocessed_df_goog['Date'])
```

### 3. Candlestick Chart (OHLC Data)

### *Explanation:*

A candlestick chart is commonly used in financial analysis to visualize price movements of a stock, commodity, or other financial instruments over time. This particular plot uses **teal** and **coral** colors to represent price increases and decreases, respectively.

#### 1. **Candlestick Components**
Candlesticks consist of four main points of data for each period (daily, weekly, etc.):
   - **Open**: The price at which the asset started trading during the period.
   - **Close**: The price at which the asset stopped trading at the end of the period.
   - **High**: The highest price reached during the period.
   - **Low**: The lowest price reached during the period.

Each candlestick has the following components:
   - **Body**: Represents the range between the **Open** and **Close** prices.
     - If the price goes **up** (Close > Open), the body is filled with **teal**.
     - If the price goes **down** (Close < Open), the body is filled with **coral**.
   - **Wicks** (or shadows): Represent the range between the **High** and **Low** prices. The wicks extend above and below the body of the candlestick.
     - The wicks for upward movements (teal) and downward movements (coral) are colored to match the candlestick body.

#### 2. **Teal and Coral Coloring**
   - **Teal**: When the closing price is higher than the opening price, it indicates that the stock price went up, so the candlestick is filled with **teal**.
   - **Coral**: When the closing price is lower than the opening price, it indicates that the stock price went down, so the candlestick is filled with **coral**.

#### 3. **50-Period Moving Average (MAV)**
   - The plot also displays a **50-period Moving Average (MAV)**. This is a line that represents the average closing price over the past 50 days. The moving average smooths out the price data to show long-term trends and helps identify potential support and resistance levels in the stock price.

#### 4. **Volume Bars**
   - The volume bars are shown below the candlestick chart. They represent the number of shares traded during the period (day, week, etc.).
   - The color of the volume bars corresponds to the direction of the price movement:
     - **Teal**: The price increased during the period (Close > Open).
     - **Coral**: The price decreased during the period (Close < Open).

#### 5. **Figure Dimensions**
   - **Figure Ratio and Scaling**: The `figratio=(12, 6)` ensures that the figure has a width of 12 units and a height of 6 units. This gives the plot a widescreen aspect ratio.
   - **Figscale=1.2**: This scales the entire plot by 1.2 times its original size, making the details easier to see.

#### 6. **Grid and Style Customization**
   - **Gridstyle**: A simple dashed line grid (`'-'`) helps guide the eye across the chart.
   - **Colors for Wick and Edge**: The wick and edge of each candlestick are colored black by default, but they can be customized to match the body of the candlestick (teal or coral).
   - **Figure Background**: The background color of the plot is set to `(0.82, 0.83, 0.85)`, which is a light greyish color, providing a neutral background for the candlestick plot.

#### 7. **Title and Labels**
   - **Title**: The title of each plot clearly states which companys data is being displayed (Google or TCS).
   - **Y-Axis Label**: The y-axis is labeled **"Price"** to indicate that its tracking the stock prices over time.
   - **X-Axis**: The x-axis represents the time (in years or other time intervals depending on your data).

---

### Why Use a Candlestick Chart?
Candlestick charts are particularly useful in stock market analysis because they visually convey essential information about price movements within a period. The **colors** help quickly identify whether a stock's price increased or decreased during the period, while the **wicks** show how volatile the price movement was (whether it traded near its high or low).

By using **teal** and **coral**, the plot effectively communicates stock price increases and decreases in a visually appealing and easy-to-understand way, making it ideal for presentations and detailed financial analysis.

### 6. Moving Average (Trend Indicator)

### *Explanation:*

A Moving Average (MA) is a statistical calculation that helps to smooth out price data by creating a continuously updated average price over a specified time period, such as 50 or 200 days. 

This smoothing effect allows traders and investors to identify the underlying trend of a stock more easily by filtering out the noise created by short-term price fluctuations. 

When the stock price is above the moving average, it generally indicates an uptrend, suggesting bullish market sentiment; conversely, when the price falls below the moving average, it indicates a downtrend, reflecting bearish sentiment. Moving averages can also serve as dynamic support and resistance levels, providing insights into potential entry and exit points for trades. 

By analyzing moving averages in conjunction with other technical indicators, traders can enhance their decision-making and gain a clearer understanding of market momentum.

```python
# Calculate moving averages for Google and TCS
goog_moving_avg = preprocessed_df_goog['Close'].rolling(window=50).mean()
tcs_moving_avg = preprocessed_df_tcs['Close'].rolling(window=50).mean()

# Plotting Moving Average for Google and TCS
fig, axs = plt.subplots(1, 2, figsize=(12, 6), facecolor = 'white')
axs[0].plot(preprocessed_df_goog.index, preprocessed_df_goog['Close'], label='Google Close', color='teal', alpha = 0.5)
axs[0].plot(preprocessed_df_goog.index, goog_moving_avg, label='Google 50-day MA', color='coral', linestyle='--')
axs[0].set_title('Google Close and Moving Average')
axs[0].set_xlabel('Time (Year)')
axs[0].set_ylabel('Price USD ($)')
axs[0].legend()

axs[1].plot(preprocessed_df_tcs.index, preprocessed_df_tcs['Close'], label='TCS Close', color='coral', alpha = 0.5)
axs[1].plot(preprocessed_df_tcs.index, tcs_moving_avg, label='TCS 50-day MA', color='teal', linestyle='--')
axs[1].set_title('TCS Close and Moving Average')
axs[1].set_xlabel('Time (Year)')
axs[1].set_ylabel('Price USD ($)')
axs[1].legend()

plt.suptitle('Moving Averages of Google and TCS', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplots_adjust(wspace=0.3)
plt.show()
```

### 10. Volatility Over Time (Rolling Standard Deviation)

### *Explanation:*

Volatility measures the extent of price fluctuations for a stock, and the rolling standard deviation provides a dynamic view of this volatility over time by calculating the standard deviation of returns within a specified window (e.g., 30 days). 

A higher rolling standard deviation indicates increased volatility, reflecting wider price variations and heightened market uncertainty, while a lower standard deviation suggests stability. 

By plotting this metric, investors can identify periods of significant market movement, assess risk, and adjust their trading strategies accordingly. This tool is especially useful for recognizing market trends, making informed decisions, and combining with other indicators to enhance investment strategies.

```python
# Calculate rolling standard deviation (30-day window) for Google and TCS
goog_rolling_std = preprocessed_df_goog['Close'].rolling(window=30).std()
tcs_rolling_std = preprocessed_df_tcs['Close'].rolling(window=30).std()

# Plotting rolling standard deviation for Google and TCS
plt.figure(figsize=(12, 6), facecolor = 'white')
plt.plot(preprocessed_df_goog.index, goog_rolling_std, label='Google Volatility', color='teal', alpha = 0.5)
plt.plot(preprocessed_df_tcs.index, tcs_rolling_std, label='TCS Volatility', color='coral', alpha = 0.5)
plt.title('Rolling Standard Deviation (Volatility): Google vs TCS')
plt.xlabel('Time (Year)')
plt.ylabel('Volatility (Std Dev)')
plt.legend()
plt.tight_layout()
plt.show()
```

### 11. Relative Strength Index (RSI): Relative Strength Index (RSI): Comparison of Google Vs TCS

### *Explanation:*

The Relative Strength Index (RSI) is a popular momentum oscillator used in technical analysis to assess the speed and change of price movements. Ranging from 0 to 100, the RSI is typically applied to a 14-day timeframe and is designed to identify overbought or oversold conditions in a stock. When the RSI value exceeds 70, it suggests that the stock may be overbought, indicating a potential price correction or pullback. 

Conversely, an RSI reading below 30 signals that the stock may be oversold, suggesting a possible rebound or upward price movement. Traders often use RSI in conjunction with other indicators to confirm trends and make informed trading decisions. Additionally, RSI can also identify divergences between the indicator and price movements, further enhancing its utility in predicting potential reversals in market direction.

```python
# Function to calculate RSI
def calculate_rsi(df, window=14):
    delta = df['Close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window).mean()
    avg_loss = loss.rolling(window).mean()
    
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    
    return rsi

# Calculating RSI for Google and TCS
goog_rsi = calculate_rsi(preprocessed_df_goog)
tcs_rsi = calculate_rsi(preprocessed_df_tcs)

# Plotting RSI for Google and TCS
fig, axs = plt.subplots(1, 2, figsize=(12, 6), facecolor = 'white')
axs[0].plot(preprocessed_df_goog.index, goog_rsi, color='teal', alpha = 0.5)
axs[0].axhline(70, color='coral', linestyle='--')
axs[0].axhline(30, color='coral', linestyle='--')
axs[0].set_title('Google RSI')
axs[0].set_xlabel('Time (Year)')
axs[0].set_ylabel('RSI')

axs[1].plot(preprocessed_df_tcs.index, tcs_rsi, color='coral', alpha = 0.5)
axs[1].axhline(70, color='teal', linestyle='--')
axs[1].axhline(30, color='teal', linestyle='--')
axs[1].set_title('TCS RSI')
axs[1].set_xlabel('Time (Year)')
axs[1].set_ylabel('RSI')

plt.suptitle('RSI Indicator: Google vs TCS', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.subplots_adjust(wspace=0.3)
plt.show()
```
